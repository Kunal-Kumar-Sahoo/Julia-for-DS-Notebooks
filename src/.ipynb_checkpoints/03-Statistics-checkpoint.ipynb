```julia
using Statistics
using StatsBase
using RDatasets
using Plots
using StatsPlots
using KernelDensity
using Distributions
using LinearAlgebra
using HypothesisTests
using PyCall
using MLBase
```

```julia
data = dataset("datasets", "faithful")
@show names(data)
data
```

```julia
describe(data)
```

```julia
eruptions = data[!, :Eruptions]
scatter(eruptions, label="eruptions")
waittime = data[!, :Waiting]
scatter!(waittime, label="wait time")
```

### 🔵Statistics plots

```julia
histogram(eruptions, label="eruptions")
```

```julia
histogram(eruptions, bins=:sqrt, label="eruptions")
```

### 🔵Kernel density estimates

```julia
p = kde(eruptions)
```

```julia
histogram(eruptions, label="eruptions")
plot!(p.x, p.density .* length(eruptions), linewidth=3, color=2, label="kde fit")
```

```julia
histogram(eruptions, bins=:sqrt, label="eruptions")
plot!(p.x, p.density .* length(eruptions) .* 0.2, linewidth=3, color=2, label="kde fit")
```

Normal Distribution

```julia
random_vector = randn(100_000)
histogram(random_vector, label="gaussian")
```

```julia
p = kde(random_vector)
plot!(p.x, p.density .* length(random_vector) .* 0.1, linewidth=3, color=2, label="kde fit")
```

### 🔵Probability distributions

```julia
d = Normal()
random_vector2 = rand(d, 100_000)
histogram(random_vector2)
p = kde(random_vector2)
plot!(p.x, p.density .* length(random_vector2) .* 0.1, linewidth=3, color=2)
```

```julia
b = Binomial(40)
random_vector3 = rand(b, 1_000_000)
histogram(random_vector3)
p = kde(random_vector3)
plot!(p.x, p.density .* length(random_vector3) .* 0.5, linewidth=3, color=2)
```

```julia
x = rand(1000)
d = fit(Normal, x)
random_vector4 = rand(d, 1000)
histogram(random_vector4, nbins=20, fillalpha=0.3, label="fit")
histogram!(x, nbins=20, linecolor=:red, fillalpha=0.3, label="vector")
```

```julia
x = eruptions
d = fit(Normal, x)
random_vector5 = rand(d, 1000)
histogram(random_vector5, nbins=20, fillalpha=0.3)
histogram!(x, nbins=20, linecolor=:red, fillalpha=0.3)
```
